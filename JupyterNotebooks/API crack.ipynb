```python
def goodReadRequest(ISBN):
    url = 'https://openlibrary.org/api/books?bibkeys=ISBN:{}&jscmd=data&format=json'.format(ISBN)
    res = requests.get(url)
    return res
```

```python
grData=[]
for i in df["isbn13"]:
    grData.append(goodReadRequest("{}".format(i)).json())
```

```python
print(len(grData))
```

```python
grDat= list(filter(None,grData))
```

```python
print(len(grDat))
```
