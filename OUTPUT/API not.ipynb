```python
import os
#from dotenv import load_dotenv
#load_dotenv()
import pandas as pd
import re 
import numpy as np
#from IPython.display import Image
#from IPython.core.display import HTML
import requests
```

```python
df = pd.read_csv('../books.csv', error_bad_lines=False)
df = df[df.average_rating > 4]    
```

```python
def goodReadRequest(ISBN):
    url = 'https://openlibrary.org/api/books?bibkeys=ISBN:{}&jscmd=data&format=json'.format(ISBN)
    res = requests.get(url)
    return res
```

```python
isbn = df.loc[1,"isbn13"]
data = goodReadRequest(isbn)
info = data.json()
info
```

```python
def getSubject(info):
    subjects = []
    publishers = []
    try:    
        for subj in list(info.values())[0]["subjects"]:
            subjects.append(subj.get("name",""))
    except:
        subjects.append(False)
    try:    
        for pub in list(info.values())[0]["publishers"]:
            publishers.append(pub.get("name",""))
    except:
        publishers.append(False)
    return subjects,publishers
```

```python
genreCol = []
publisherCol = []
for index, _ in df.iterrows():
    isbn = df.loc[index,"isbn13"]
    data = goodReadRequest(isbn)
    info = data.json()
    sub, pub = getSubject(info)
    genreCol.append(sub)
    publisherCol.append(pub)
    
```

```python
df["Genre"] = genreCol
df["Publisher"]=publisherCol
df.to_csv("dfinal.csv")
```
